```python
from explain_evaluator import VerificationNetwork
from transformers import AutoTokenizer
import torch

%load_ext autoreload
%autoreload 2
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
explainer = VerificationNetwork('bert-base-cased', tokenizer.mask_token_id)
```

```python
premise = "Your gift is appreciated by each and every student who will benefit from your generosity."	
hypothesis = "Hundreds of students will benefit from your generosity."

inputs = tokenizer(premise, text_pair=hypothesis, return_tensors='pt')

y = torch.tensor([1])
```

```python
out = explainer(**inputs)
```

```python
explainer.backward(out, y)
```
